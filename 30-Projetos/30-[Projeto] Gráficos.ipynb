Vamos analisar dados sobre o _gap_ salarial entre homens e mulheres no Reino Unido. Os dados são provenientes do _Anual Survey of Earnings and Hours_ de 2022 e estão disponíveis [aqui]('https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/data/2022/2022-06-28/UK%20Gender%20Pay%20Gap%20Data%20-%202022%20to%202023.csv'). Originalmente, eles foram disponibilizados aqui [gender-pay-gap.service.gov.uk](https://gender-pay-gap.service.gov.uk/viewing/download), junto com um [blog e dashboard](https://www.ons.gov.uk/employmentandlabourmarket/peopleinwork/earningsandworkinghours/articles/findoutthegenderpaygapforyourjob/2016-12-09) para você explorar os dados.

Rode o seguinte código para carregar os dados. Nós utilizaremos apenas algumas colunas. Uma delas é categórica ordinal, de modo que teremos que colocar suas categorias em ordem. Tudo isso é feito no código a seguir:

````python
paygap = pd.read_csv('https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/data/2022/2022-06-28/UK%20Gender%20Pay%20Gap%20Data%20-%202022%20to%202023.csv',
                    usecols=['EmployerName',
                             'DiffMedianHourlyPercent',
                             'DiffMedianBonusPercent',
                             'MaleBonusPercent',
                             'FemaleBonusPercent',
                             'EmployerSize'], index_col='EmployerName')

paygap['EmployerSize'] = pd.Categorical(paygap.EmployerSize, categories=["Less than 250", "250 to 499", "500 to 999","1000 to 4999","5000 to 19,999"], ordered=True)

paygap.head(5)
````

* `DiffMedianHourlyPercent` é a mediana da diferença percentual entre a remuneração pela hora trabalhada por homens e mulheres (positivo = valor mediano da hora do homem é mais elevado)
* `DiffMedBonusPercent` é a mediana da diferença percentual entre o bônus de homens e mulheres (positivo = valor mediano do bônus do homem é mais elevado)
* `MaleBonusPercent` e `FemaleBonusPercent` são as porcentagens das remunerações pagas a título de bônus para homens e mulheres.
* `EmployerSize` é a faixa de tamanho da empresa, medido em termos do número de empregados.

Cada linha representa uma empresa.
